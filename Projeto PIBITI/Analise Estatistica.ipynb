```python
# Esse laço entra no vetor estacões
for estacao in estacoes:
    print('Trabalhando <', end = '')
    #Entra em cada estação e ler suas tabelas
    pressao = pd.read_excel('Banco de dados Analise/%s' % estacao[0])
    temperatura = pd.read_excel('Banco de dados Analise/%s' % estacao[1])
    
    #Nome da estação
    nome = estacao[0]
    for i in reversed(range(len(nome))):
        if nome[i] == '_':
            break;
    nome = nome[0:i]
    
    #Exclui linha das horas
    pressao = pressao.drop(0)
    temperatura = temperatura.drop(0)
    
    #error 
    # 1 = Pressão atmosférica : média aritmética
    # 2 = Radiação global : soma dos valores positivos
    # 3 = Temperatura do ar : média aritmética
    # 4 = TEMPERATURA MAXIMA (°C)
    # 5 = (TPmine - TPmax) valor negativo, invalidando  raiz  em ET0HS
    # 6 = Velocidade do vento : média aritmética
    err = 0
    errs = []
    dia_invalido = 0
    
    #Conta a quantidade de linhas de cada tabela
    pressao_nrows = len(pressao.index)
    temperatura_nrows = len(temperatura.index)
    
    dia = []
    medias_pressoes = []
    medias_vento = []
    soma_radiacoes = []
    media_temperaturas = []
    media_umidades = []
    medias_orvalhos = []
    dias_julianos = []
    tpmaxs = []
    tpmins = []
    dias_et0PM = []
    dias_et0HS = []
    dias_et0IV = []
    errs = []

    latitude = pressao['LATITUDE'][1]
    latitude = np.radians(latitude)
    altitude = pressao['ALTITUDE (m)'][1]

    for i in range(1, pressao_nrows+1, 1): #linhas / Dia
        err = 0
        dia_pressao = []        
        dia_velocidade = []         
        dia_radiacao = []
        dia_temperatura = []
        dia_umidade = []
        dia_orvalho = []
        for j in range(24): #Colunas
            #Pressão atmosférica : média aritmética
            if j == 0:
                dia_pressao.append(pressao['PRESSÃO ATMOSFERICA (hPa)'][i])
            else:
                dia_pressao.append(pressao['PRESSÃO ATMOSFERICA (hPa).%i' % j ][i])
        
            #Velocidade do vento : média aritmética
            if j== 0:
                dia_velocidade.append(pressao['VENTO VELOCIDADE (m/s)'][i])
            else:
                dia_velocidade.append(pressao['VENTO VELOCIDADE (m/s).%i' % j ][i])
            
            #Radiação global : soma dos valores positivos
            if j <= 13:
                if j == 0:
                    if ((pressao['RADIACAO GLOBAL (KJ/M²)'][i]) >= 0):
                        dia_radiacao.append(pressao['RADIACAO GLOBAL (KJ/M²)'][i])
                else:
                    if (pressao['RADIACAO GLOBAL (KJ/M²).%i' % j ][i]) >= 0:
                        dia_radiacao.append(pressao['RADIACAO GLOBAL (KJ/M²).%i' % j ][i])

            #Temperatura do ar : média aritmética
            if j == 0:
                dia_temperatura.append(temperatura['TEMPERATURA DO AR (°C)'][i])
            else:
                dia_temperatura.append(temperatura['TEMPERATURA DO AR (°C).%i' % j ][i])

            #Umidade relativa do ar : média aritmética
            if j == 0:
                dia_umidade.append(temperatura['UMIDADE RELATIVA DO AR (%)'][i])
            else:
                dia_umidade.append(temperatura['UMIDADE RELATIVA DO AR (%%).%i' % j ][i])

            #Temperatura de ponto de orvalho : média aritmética
            if j == 0:
                dia_orvalho.append(temperatura['TEMPERATURA DO PONTO DE ORVALHO (°C)'][i])
            else:
                dia_orvalho.append(temperatura['TEMPERATURA DO PONTO DE ORVALHO (°C).%i' % j ][i])

        #Temperatura maxima
        tpmax = temperatura['TEMPERATURA MAXIMA (°C)'][i]
            
        tpmin = temperatura['TEMPERATURA MINIMA (°C)'][i]

        #tpminPO = temperatura['TEMPERATURA MÁXIMA DO PONTO DE ORVALHO (°C)'][i]

        #tpmaxPO = temperatura['TEMPERATURA MÍNIMA DO PONTO DE ORVALHO (°C)'][i]

        umax = temperatura['UMIDADE RELATIVA MAXIMA DO AR (%)'][i]

        umin = temperatura['UMIDADE RELATIVA MINIMA DO AR (%)'][i]
        
        media_dia_pressao = 0   
        media_dia_velocidade = 0  
        soma_dia_radiacao = 0
        media_dia_temperatura = 0
        media_dia_umidade = 0
        media_dia_orvalho = 0
        dia_juliano = pressao['DIA JULIANO'][i]
        media_dia_pressao = np.mean(dia_pressao) / 10
        media_dia_velocidade = np.mean(dia_velocidade)
        soma_dia_radiacao = np.sum(dia_radiacao) / 1000 
        media_dia_temperatura = np.mean(dia_temperatura)   
        media_dia_umidade = np.mean(dia_umidade)
        media_dia_orvalho = np.mean(dia_orvalho)
        
        #Verificação  de erros
        if (media_dia_pressao <= 30) | (media_dia_pressao == None):
            err = 1
        elif (soma_dia_radiacao <= 8) | (soma_dia_radiacao == None):
            err = 2
        elif (media_dia_temperatura <= 5) | (media_dia_temperatura == None):
            err = 3
        elif (tpmax <= 5) | (tpmax == None):
            err = 4 
        elif (tpmax - tpmin) <= 0 :
            err = 5
        elif (media_dia_velocidade <= 0) | (media_dia_velocidade == None):
            err = 6
            
        # Calculo
        if err == 0:
            e0max = (6.1078 * (10 ** ( (7.5*tpmax) / (237.3+tpmax) ) ) ) / 10

            e0min = (6.1078 * (10 ** ( (7.5*tpmin) / (237.3+tpmin) ) ) ) / 10

            es = (e0min + e0max) / 2

            ea = ( (e0min * (umax / 100) ) + (e0max * (umin / 100) ) ) / 2

            u2 = media_dia_velocidade * ( (np.log( (2 - 0.08) / (0.01476) ) ) / np.log( (10 - 0.08) / (0.01476)) )

            delta = ( (4098 * es) / (  (media_dia_temperatura + 237.3) ** 2) )

            teste_pressao = ( 1013.25 - (0.106 * altitude) ) / 10

            gama = ( (0.0016286 * media_dia_pressao) / 2.45)

            inclinacao = np.radians(23.45) * np.sin (  ( (np.radians(360) / 365) * (284 + dia_juliano) )  )

            hn = np.arccos(  ( -np.tan(latitude) ) * np.tan(inclinacao) )

            radAtm = 37.6 * ( ( hn *  np.sin(latitude) * np.sin(inclinacao) ) + ( np.cos(latitude) * np.cos(inclinacao) * np.sin(hn) )  )

            rs = (0.75 + ( 2 * (10  ** -5) ) * 280 ) * radAtm

            rol = 4.9 * (10 ** -9) * ( ( ( ( (tpmax + 273)  ** 4) ) + ( (tpmin + 273) ** 4)  ) / 2 ) * (  0.34 - (0.14 * (np.sqrt(ea) ) ) )  * (1.35 * (soma_dia_radiacao / rs) - 0.35) 

            rn = soma_dia_radiacao -(0.23  * soma_dia_radiacao) - rol

            et0PM = (((0.408  *  delta * rn) + gama * (900 / (media_dia_temperatura + 273)) * u2  * (es - ea)) / (delta + (gama * (1 + 0.34 * u2) ) ) )
            
            et0HS = 0.0023 * (radAtm  / 2.45) * (np.sqrt(tpmax - tpmin)) * (media_dia_temperatura + 17.8)
                
            et0IV = 0.006 * ( ( (25 + media_dia_temperatura)** 2) * (1 - (media_dia_umidade/ 100)))
            
            
        else:
            et0PM = None
            et0HS = None
            et0IV = None
            dia_invalido += 1
        
        dia.append(temperatura['Unnamed: 0'][i].strftime('%d/%m/%Y'))
        medias_pressoes.append(media_dia_pressao)
        medias_vento.append(media_dia_velocidade)
        soma_radiacoes.append(soma_dia_radiacao)
        media_temperaturas.append(media_dia_temperatura)
        media_umidades.append(media_dia_umidade)
        medias_orvalhos.append(media_dia_orvalho)
        dias_julianos.append(dia_juliano)
        tpmaxs.append(tpmax)
        tpmins.append(tpmin)
        dias_et0PM.append(et0PM)
        dias_et0HS.append(et0HS)
        dias_et0IV.append(et0IV)
        errs.append(err)  
        
        # Controle de tempo
        if i == int(pressao_nrows / 4):
            print('...25%...', end = '')
        elif i == int(pressao_nrows / 2):
            print('...50%...', end = '')
        elif i == int(pressao_nrows - ((pressao_nrows / 2) / 2)):
            print('...75%...', end = '')      
    
    #Criando documento da estação com os resultados
    dataFrame = pd.DataFrame({'Dia' : dia,
                               'Média Pressão atmosférica' : medias_pressoes,
                               'Média Velocidade do vento' : medias_vento,
                               'Radiacão global soma dos valores positivos' : soma_radiacoes,
                               'Média Temperatura do ar' : media_temperaturas,
                               'Média Umidade relativa do ar' : media_umidades,
                               'Média Temperatura de ponto de orvalho' : medias_orvalhos,
                               'TEMPERATURA MAXIMA (°C)' : tpmaxs,
                               'TEMPERATURA MINIMA (°C)' : tpmins,
                               'DIA JULIANO' : dias_julianos,
                               'LATITUDE' : latitude,
                               'ALTITUDE (m)' : altitude,
                               'Evapotranspiração PENMAN-MONTEITH FAO56' : dias_et0PM,
                               'Evapotranspiração HARGREAVES-SAMANI' : dias_et0HS,
                               'Evapotranspiração IVANOV' : dias_et0IV,
                               'ERROR' : errs,})
    
    folha = {nome : dataFrame}
          
    #Validar quantidade de anos <= 3
    if (pressao_nrows  - dia_invalido) < 1096:
        print("> Estação %s sem dados suficientes!" % nome)
        continue
        
    #Gerando tabela com  os resultados
    excel = pd.ExcelWriter('Resultados Analise/ET0s/%s.xlsx'  % nome, engine='xlsxwriter',
                      date_format='YYYY-MM-DD',
                      datetime_format='YYYY-MM-DD HH:MM:SS')
    
    for sheet_name in folha.keys():
        folha[sheet_name].to_excel(excel, sheet_name=sheet_name, index=False)
        workbook  = excel.book
        worksheet = excel.sheets[sheet_name]

        # Adicionar um formato de cabeçalho.
        header_format = workbook.add_format({
            'bold': True,
            'text_wrap': True,
            'valign': 'top',
            'fg_color': '#D7E4BC',
            'border': 1})

        # Escrevendo os cabeçalhos das colunas com o formato definido.
        for col_num, value in enumerate(folha[sheet_name].columns.values):
            worksheet.write(0, col_num, value, header_format)

    # Salvando documento!
    excel.save()

    print("> Estação %s concluída!" % nome)
```

```python
et0s = os.listdir('Resultados Analise/ET0s/')
# Esse laço entra no vetor estacões
for et0 in et0s:
    #Entra em cada estação e ler suas tabelas
    estacao = pd.read_excel('Resultados Analise/ET0s/%s' % et0)
    
    nome = et0
    nome = nome[0:len(nome)-5]
    
    #Linhas
    estacao_nrows = len(estacao.index)

    et0PMdia = [[]]*365
    et0IVdia = [[]]*365
    et0HSdia = [[]]*365
    dp_et0PM = [0]*365
    dp_et0IV = [0]*365
    dp_et0HS = [0]*365
    ivpm1dia = []
    hspm1dia = []
    ivpm2dia = []
    hspm2dia = []
    epeIVdia = 0
    dIVdia = 0
    epeHSdia = 0
    dHSdia = 0
    dias = [0]*365
    janIV = 0
    fevIV = 0
    macIV = 0
    abrIV = 0
    maiIV = 0
    junIV = 0
    julIV = 0
    agoIV = 0
    setIV = 0
    outIV = 0
    novIV = 0
    dezIV = 0
    mediaIV = 0
    janHS = 0
    fevHS = 0
    macHS = 0
    abrHS = 0
    maiHS = 0
    junHS = 0
    julHS = 0
    agoHS = 0
    setHS = 0
    outHS = 0
    novHS = 0
    dezHS = 0
    mediaHS = 0
    janPM = 0
    fevPM = 0
    macPM = 0
    abrPM = 0
    maiPM = 0
    junPM = 0
    julPM = 0
    agoPM = 0
    setPM = 0
    outPM = 0
    novPM = 0
    dezPM = 0
    mediaPM = 0
    
    for i in range(365):
        et0PMdia[i] = []
        et0IVdia[i] = []
        et0HSdia[i] = []
    
    for i in range(estacao_nrows): #linhas / Dia
        dia = int(estacao['DIA JULIANO'][i])
        dias[dia-1] = dia
        if np.isnan(estacao['Evapotranspiração PENMAN-MONTEITH FAO56'][i]) != True:
            et0PMdia[dia-1] += [estacao['Evapotranspiração PENMAN-MONTEITH FAO56'][i]]
            
        if np.isnan(estacao['Evapotranspiração IVANOV'][i]) != True:
            et0IVdia[dia-1] += [estacao['Evapotranspiração IVANOV'][i]]
            
        if np.isnan(estacao['Evapotranspiração HARGREAVES-SAMANI'][i]) != True:
            et0HSdia[dia-1] += [estacao['Evapotranspiração HARGREAVES-SAMANI'][i]]
            
    for i in range(365):
        dp_et0PM[i] = np.std(et0PMdia[i])
        dp_et0IV[i] = np.std(et0IVdia[i])
        dp_et0HS[i] = np.std(et0HSdia[i])
        et0PMdia[i] = np.mean(et0PMdia[i])
        et0IVdia[i] = np.mean(et0IVdia[i])
        et0HSdia[i] = np.mean(et0HSdia[i])
        ivpm1dia.append(( et0PMdia[i] - et0IVdia[i] )**2)
        hspm1dia.append(( et0PMdia[i] - et0HSdia[i] )**2)    
    
    mediaPMdia = np.mean(et0PMdia)
    
    for i in range(365):
        ivpm2dia.append(((abs(et0IVdia[i] - mediaPMdia)) + (abs(et0PMdia[i] - mediaPMdia)))**2)
        hspm2dia.append(((abs(et0HSdia[i] - mediaPMdia)) + (abs(et0PMdia[i] - mediaPMdia)))**2)
        
        
    epeIVdia =  np.sqrt(np.sum(ivpm1dia) / (365-1))
    dIVdia = 1 - (np.sum(ivpm1dia) / np.sum(ivpm2dia))
    
    epeHSdia =  np.sqrt(np.sum(hspm1dia) / (365-1))
    dHSdia = 1 - (np.sum(hspm1dia) / np.sum(hspm2dia))
    
    #Médias et0IV
    janIV = np.mean(et0IVdia[0:30])
    fevIV = np.mean(et0IVdia[31:58])
    macIV = np.mean(et0IVdia[59:89])
    abrIV = np.mean(et0IVdia[90:119])
    maiIV = np.mean(et0IVdia[120:150])
    junIV = np.mean(et0IVdia[151:180])
    julIV = np.mean(et0IVdia[181:211])
    agoIV = np.mean(et0IVdia[212:242])
    setIV = np.mean(et0IVdia[243:272])
    outIV = np.mean(et0IVdia[273:303])
    novIV = np.mean(et0IVdia[304:333])
    dezIV = np.mean(et0IVdia[334:])
    aux = [janIV, fevIV, macIV,  abrIV, maiIV, junIV, julIV, agoIV, setIV, outIV, novIV, dezIV]
    mediaIV = np.mean(aux)

    #Médias et0HS
    janHS = np.mean(et0HSdia[0:30])
    fevHS = np.mean(et0HSdia[31:58])
    macHS = np.mean(et0HSdia[59:89])
    abrHS = np.mean(et0HSdia[90:119])
    maiHS = np.mean(et0HSdia[120:150])
    junHS = np.mean(et0HSdia[151:180])
    julHS = np.mean(et0HSdia[181:211])
    agoHS = np.mean(et0HSdia[212:242])
    setHS = np.mean(et0HSdia[243:272])
    outHS = np.mean(et0HSdia[273:303])
    novHS = np.mean(et0HSdia[304:333])
    dezHS = np.mean(et0HSdia[334:])
    aux = [janHS, fevHS, macHS,  abrHS, maiHS, junHS, julHS, agoHS, setHS, outHS, novHS, dezHS]
    mediaHS = np.mean(aux)

    #Médias et0PM
    janPM = np.mean(et0PMdia[0:30])
    fevPM = np.mean(et0PMdia[31:58])
    macPM = np.mean(et0PMdia[59:89])
    abrPM = np.mean(et0PMdia[90:119])
    maiPM = np.mean(et0PMdia[120:150])
    junPM = np.mean(et0PMdia[151:180])
    julPM = np.mean(et0PMdia[181:211])
    agoPM = np.mean(et0PMdia[212:242])
    setPM = np.mean(et0PMdia[243:272])
    outPM = np.mean(et0PMdia[273:303])
    novPM = np.mean(et0PMdia[304:333])
    dezPM = np.mean(et0PMdia[334:])
    aux = [janPM, fevPM, macPM,  abrPM, maiPM, junPM, julPM, agoPM, setPM, outPM, novPM, dezPM]
    mediaPM = np.mean(aux)

    
    def grafico(x, nome_metodo):
        y = np.array(et0PMdia)
        y = y.reshape(-1,1)
        x = x.reshape(-1,1)
        regressor =  LinearRegression()
        regressor.fit(x,y)
        plt.scatter(x,y)
        plt.plot(x, regressor.predict(x), color = 'red')
        plt.title(' ')
        plt.ylabel('ETo Penman-Monteith (mm/dia)')
        plt.xlabel('%s (mm/dia)'% nome_metodo)
        plt.savefig('Resultados Analise/medias finais/Graficos/%s_%s.png' % \
                    (nome, nome_metodo), bbox_inches='tight')
        plt.close()
        
    y = np.array(et0PMdia)
    
    def to_df(aux):
        model_results_df = []
        for i, tab in enumerate(aux.summary().tables):
            header, index_col = None, None
            if i == 1:
                coefficient_df = pd.read_html(tab.as_html(), header=0, index_col=0)[0]
            else:
                df = pd.read_html(tab.as_html())[0]
                model_results_df += [df.iloc[:,0:2], df.iloc[:,2:4]]

        model_results_df = pd.DataFrame(np.concatenate(model_results_df), columns=['metric', 'value'])
        model_results_df.dropna(inplace=True, axis=0)
        model_results_df.metric = model_results_df.metric.apply(lambda x : x.lower().replace(' (', '_')
                                                        .replace('.', '').replace('(', '_')
                                                        .replace(')', '').replace('-', '_')
                                                       .replace(':', '').replace(' ', '_'))

        res_dict = pd.DataFrame(zip(model_results_df.metric.values, model_results_df.value.values))
        df1 = (aux.summary2().tables[1])
        return pd.concat([df1, res_dict], ignore_index=False, sort=False)
    
    #como ter os resultado da tabela IVPM
    x = np.array(et0IVdia)
    resultsIVPM = sm.OLS(y, sm.add_constant(x)).fit()
    grafico(x, "ETo Ivanov")
    dfIV = to_df(resultsIVPM)
    
    #como ter os resultado da tabela HSPM
    x = np.array(et0HSdia)
    resultsHSPM = sm.OLS(y, sm.add_constant(x)).fit()
    grafico(x, "ETo Hargreaves-Samani")
    dfHS = to_df(resultsHSPM)
    
    folha = {"ET0IV" : dfIV,
             "ET0HS"  : dfHS}
          
    #Gerando tabela com  os resultados
    excel = pd.ExcelWriter('Resultados Analise/medias finais/%s.xlsx'  % nome, engine='xlsxwriter',
                      date_format='YYYY-MM-DD',
                      datetime_format='YYYY-MM-DD HH:MM:SS')
    
    for sheet_name in folha.keys():
        folha[sheet_name].to_excel(excel, sheet_name=sheet_name, index=False)
        workbook  = excel.book
        worksheet = excel.sheets[sheet_name]

        # Adicionar um formato de cabeçalho.
        header_format = workbook.add_format({
            'bold': True,
            'text_wrap': True,
            'valign': 'top',
            'fg_color': '#D7E4BC',
            'border': 1})
        
        worksheet.write(1, 1, 'const')
        worksheet.write(2, 1, 'x1')
        
        worksheet.write(5, 7, 'EPE IV')
        worksheet.write(6, 7, epeIVdia)
        worksheet.write(7, 7, 'D IV')
        worksheet.write(8, 7, dIVdia)
            
        worksheet.write(5, 6, 'EPE HS')
        worksheet.write(6, 6, epeHSdia)
        worksheet.write(7, 6, 'D HS')
        worksheet.write(8, 6, dHSdia)
        
        worksheet.write(1, 9, 'JAN')
        worksheet.write(2, 9, 'FEV')
        worksheet.write(3, 9, 'MAÇ')
        worksheet.write(4, 9, 'ABR')
        worksheet.write(5, 9, 'MAI')
        worksheet.write(6, 9, 'JUN')
        worksheet.write(7, 9, 'JUL')
        worksheet.write(8, 9, 'AGO')
        worksheet.write(9, 9, 'SET')
        worksheet.write(10, 9, 'OUT')
        worksheet.write(11, 9, 'NOV')
        worksheet.write(12, 9, 'DEZ')
        worksheet.write(13, 9, 'Média Final')
        worksheet.write(0, 11, 'Média IV')
        worksheet.write(1, 11, janIV)
        worksheet.write(2, 11, fevIV)
        worksheet.write(3, 11, macIV)
        worksheet.write(4, 11, abrIV)
        worksheet.write(5, 11, maiIV)
        worksheet.write(6, 11, junIV)
        worksheet.write(7, 11, julIV)
        worksheet.write(8, 11, agoIV)
        worksheet.write(9, 11, setIV)
        worksheet.write(10, 11, outIV)
        worksheet.write(11, 11, novIV)
        worksheet.write(12, 11, dezIV)
        worksheet.write(13, 11, mediaIV)
        worksheet.write(0, 12, 'Média HS')
        worksheet.write(1, 12, janHS)
        worksheet.write(2, 12, fevHS)
        worksheet.write(3, 12, macHS)
        worksheet.write(4, 12, abrHS)
        worksheet.write(5, 12, maiHS)
        worksheet.write(6, 12, junHS)
        worksheet.write(7, 12, julHS)
        worksheet.write(8, 12, agoHS)
        worksheet.write(9, 12, setHS)
        worksheet.write(10, 12, outHS)
        worksheet.write(11, 12, novHS)
        worksheet.write(12, 12, dezHS)
        worksheet.write(13, 12, mediaHS)
        worksheet.write(0, 10, 'Média PM')
        worksheet.write(1, 10, janPM)
        worksheet.write(2, 10, fevPM)
        worksheet.write(3, 10, macPM)
        worksheet.write(4, 10, abrPM)
        worksheet.write(5, 10, maiPM)
        worksheet.write(6, 10, junPM)
        worksheet.write(7, 10, julPM)
        worksheet.write(8, 10, agoPM)
        worksheet.write(9, 10, setPM)
        worksheet.write(10, 10, outPM)
        worksheet.write(11, 10, novPM)
        worksheet.write(12, 10, dezPM)
        worksheet.write(13, 10, mediaPM)
        
        worksheet.write(10, 4, 'R')
        worksheet.write_formula('E12', '=SQRT(B14:B14)')
        
        worksheet.set_column('A:A', 20)
        worksheet.set_column('B:B', 20)
    
        
    # Salvando documento!
    excel.save()
    
    print ('Concluido {}'.format(et0))
```
